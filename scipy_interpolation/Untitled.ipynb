```python
import scipy.interpolate as interpolate
import numpy as np
def f(x): return 0.5*x + np.sin(x) 
x0 = np.linspace(1.0, 3.0, 5)
a = interpolate.splrep(x0, f(x0), k=1) # B-spline representation of 1-D curve
x = np.linspace(1.0, 3.0, 50)
y = interpolate.splev(x, a) # Evaluate a B-spline or its derivatives
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x0, f(x0), 'o', markersize=10)
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(x, y, 'r--', label='interpolation')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
